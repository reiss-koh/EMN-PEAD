```python
import numpy as np
import pandas as pd
import datetime
from dateutil.parser import parse
```

```python
df = pd.read_excel('/content/drive/MyDrive/FH_processing.xlsx')
symbol_list = list(df['Symbol'])
symbol_list
```

```python
df1 = pd.read_excel('/content/drive/MyDrive/E[R]_5D_python.xlsx')
df1
```

```python
condition_list = []
for i in df1['Ticker']:
  if i in symbol_list:
    condition_list.append(True)
  else:
    condition_list.append(False)
df1 = df1[condition_list]
df1.reset_index(drop = True, inplace = True)
df1
```

```python
df2 = pd.read_excel('/content/drive/MyDrive/E[R]_5D_python1.xlsx')
df2
```

```python
condition_list = []
for i in df2['Ticker']:
  if i in symbol_list:
    condition_list.append(True)
  else:
    condition_list.append(False)
df2 = df2[condition_list]
df2.reset_index(drop = True, inplace = True)
df2
```

```python
d1 = datetime.timedelta(days = 1)
d2 = datetime.timedelta(days = 2)
for i in range(df2.shape[0]):
  for j in range(2, df2.shape[1]):
    if type(df2.iloc[i, j]) == datetime.datetime:
      if df1.iloc[i, j] == 1:
        df2.iloc[i, j] = df2.iloc[i, j] + d1
      elif df1.iloc[i, j] == 7:
        df2.iloc[i, j] = df2.iloc[i, j] + d2
    elif type(df2.iloc[i, j]) == str:
      df2.iloc[i, j] = parse(df2.iloc[i, j])
      if df1.iloc[i, j] == 1:
        df2.iloc[i, j] = df2.iloc[i, j] + d1
      elif df1.iloc[i, j] == 7:
        df2.iloc[i, j] = df2.iloc[i, j] + d2
df2
```

```python
df2.to_excel('/content/drive/MyDrive/E[R]_5D_processing.xlsx')
```
