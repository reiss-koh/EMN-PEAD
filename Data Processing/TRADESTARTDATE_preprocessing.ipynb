## TRADESTARTDATE

```python
df = pd.read_excel('/content/drive/MyDrive/ALL_TRADESTARTDATE_python.xlsx')
df
```

```python
diff_days = datetime.timedelta(days = 1)
for i in list(df.columns[2:]):
  for j, day in enumerate(df[i]):
    while type(day) == type(df['FQ12012'][0]):
      if (day not in kor_holidays) and (day.weekday() in [0, 1, 2, 3, 4]):
        break
      day = day + diff_days
    df[i][j] = day
df
```

```python
df.to_excel('/content/drive/MyDrive/ALL_TRADESTARTDATE1_READY.xlsx')
```

```python
diff_days = datetime.timedelta(days = 1)
for e in [5, 10, 20, 30, 40, 50]:
  df0 = df.copy()
  for i in list(df0.columns[2:]):
    for j, day in enumerate(df0[i]):
      if type(day) == type(df0['FQ12012'][0]):
        for _ in range(e):
          day = day + diff_days
          while (day.weekday() in [5, 6]) or (day in kor_holidays):
            day = day + diff_days
        df0[i][j] = day
  df0.to_excel(f'/content/drive/MyDrive/ALL_TRADESTARTDATE{e}_processing.xlsx')
```
