## SUEBIT

```python
for i in list(df1.columns[2:]):
  Q3 = df1[i].quantile(q = 0.75)
  Q1 = df1[i].quantile(q = 0.25)
  IQR= Q3 - Q1
  condition = (df1[i] > (Q3 + IQR * 1.5)) | (df1[i] < (Q1 - IQR * 1.5))
  df1[i][condition] = np.NaN
df1
```

```python
sns.boxplot(data = df1.iloc[:, 2:])
```

```python
sns.histplot(data = df1.iloc[:, 2:], legend = False)
```

```python
fig = plt.figure(figsize = (15, 65))
for i, e in enumerate(list(df1.columns[2:])):
  df_list = df1[e].dropna()
  df_list = np.array(df_list, dtype = 'float32')
  ax = fig.add_subplot(13, 3, i + 1)
  stats.probplot(df_list, dist = stats.norm, plot = ax)
plt.show()
```
